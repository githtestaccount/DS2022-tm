esoph {datasets}	R Documentation
Smoking, Alcohol and (O)esophageal Cancer

### Description

Data from a case-control study of (o)esophageal cancer in Ille-et-Vilaine, France.

### Format

A data frame with records for 88 age/alcohol/tobacco combinations.

1. "agegp" - Age group 	
1. "alcgp" - Alcohol consumption
1. "tobgp" - Tobacco consumption
1. "ncases" - Number of cases 	
1. "ncontrols" - Number of controls 	

### Author(s)

Thomas Lumley

### Source

Breslow, N. E. and Day, N. E. (1980) Statistical Methods in Cancer Research. Volume 1: The Analysis of Case-Control Studies. IARC Lyon / Oxford University Press.

```r
plot(esoph$ncases, esoph$ncontrols, xlab="Cases", ylab="Controls", main="Simple Plot of Cases vs Controls")
```

```r
stripchart(ncases ~ tobgp, data=esoph, main="Tobacco Consumption vs Number of Cases")
```

```r
stripchart(ncases ~ alcgp, data=esoph, main="Alcohol Consumption vs Number of Cases")
```
