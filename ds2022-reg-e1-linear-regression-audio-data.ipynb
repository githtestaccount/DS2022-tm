## Linear regression with audio data
no sensible use case in the way it's dobe here, just for experimenting with an unusual form of data

```python
# read in a random 'dirty' audio files from http://www-mmsp.ece.mcgill.ca/Documents/AudioFormats/WAVE/Samples.html (first google hit)

url = "http://www-mmsp.ece.mcgill.ca/Documents/AudioFormats/WAVE/Samples/AFsp/M1F1-int24WE-AFsp.wav"
data, sr = sf.read(io.BytesIO(urlopen(url).read()))
```

```python
ipd.Audio("http://www-mmsp.ece.mcgill.ca/Documents/AudioFormats/WAVE/Samples/AFsp/M1F1-int24WE-AFsp.wav")
```

```python
# get a peek

print(f'data: {data[:10]}')
print(f'shape data: {data.shape}')
print(f'samplerate: {sr}')
```

```python
# get first channel pandas version

ch0 = pd.Series(data[:,0])
ch0.plot(figsize=(10, 5), lw=1, title='Channel 0')
plt.show()
```

```python
# select 200 frames that are hopefully slightly off means

snippet = ch0[1000:1200]
snippet.plot(figsize=(10, 5), lw=1, title='200 Frames Snippet from Channel 0')
plt.show()
snippet.describe()
```

```python
# rehshape data for sklearn.linear_model.LinearRegression

# frameindex = np.arange(200)
frameindex = snippet.index
x = np.reshape(frameindex, (-1, 1))
y = snippet.values
```

```python
# testing/validation only for completeness of the process

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/7, random_state = 0)
```

```python
# fit model

model = LinearRegression().fit(x_train, y_train)
y_prediction = model.predict(x_test)
print(y_prediction)
```

```python
# plot test data

plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, model.predict(x_train), color = 'blue')
plt.title('Test Data')
plt.show()
print("Mean: ", model.predict(x_test).mean())
```

```python
# plot prediction

snippet.plot(figsize=(10, 5), lw=1, title='Prediction for Full Data Set')
plt.plot(x, model.predict(x), color = 'red')
plt.show()
```

```python
# close call

print(snippet.mean())
print(model.predict(x).mean())
```

## with a small sample (even more invalid use case, just to see the off-means effect)

```python
# select 20 frames that will be more off means

smaller_snippet = ch0[1020:1040]
smaller_snippet.plot(figsize=(10, 5), lw=1, title='X Frames Snippet from Channel 0')
plt.show()
smaller_snippet.describe()
```

```python
# rehshape data for sklearn.linear_model.LinearRegression

# frameindex = np.arange(200)
s_frameindex = smaller_snippet.index
x2 = np.reshape(s_frameindex, (-1, 1))
y2 = smaller_snippet.values
```

```python
s_xTrain, s_xTest, s_yTrain, s_yTest = train_test_split(x2, y2, test_size = 1/7, random_state = 0)
model = LinearRegression().fit(s_xTrain, s_yTrain)
s_yPrediction = model.predict(s_xTest)
print(s_yPrediction)
```

```python
# plot test data

plt.scatter(s_xTest, s_yTest, color = 'red')
plt.plot(s_xTrain, model.predict(s_xTrain), color = 'blue')
plt.title('Test Data (Smaller Sample)')
plt.show()
print("Mean: ", model.predict(s_xTest).mean())
```

```python
# plot prediction

smaller_snippet.plot(figsize=(10, 5), lw=1, title='Smaller Sample Prediction')
plt.plot(x2, model.predict(x2), color = 'red')
plt.show()
```
